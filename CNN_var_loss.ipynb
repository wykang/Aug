```python
'''Convert class labels from scalars to one-hot vectors.'''
def dense_to_one_hot(labels_dense, num_classes=10):
    num_labels = labels_dense.shape[0]
    index_offset = np.arange(num_labels) * num_classes
    labels_one_hot = np.zeros((num_labels, num_classes), dtype=np.float32)
    labels_one_hot.flat[index_offset + labels_dense.ravel()] = 1
    return labels_one_hot
```

```python
def calc_moments(x, mode):
    batch_mean, batch_var = tf.nn.moments(x, [0], name='moments')
    ema = tf.train.ExponentialMovingAverage(decay=0.9)
    ema_apply_op = ema.apply([batch_mean, batch_var])
    ema_mean, ema_var = ema.average(batch_mean), ema.average(batch_var)
    def mean_var_with_update():
                with tf.control_dependencies([ema_apply_op]):
                    return tf.identity(batch_mean), tf.identity(batch_var)
    mean, var = control_flow_ops.cond(mode,
        mean_var_with_update,
        lambda: (ema_mean, ema_var))
    return mean, var
```

```python
def var_loss_fc(features, labels, idx, var_name):
    locs = tf.where(tf.equal(labels, idx))
    batch_mean, batch_var = tf.nn.moments(tf.gather(features, locs), [0], name=var_name)
    return batch_mean, batch_var
```

```python
def center_loss(features, labels, idx, var_name):
    locs = tf.where(tf.equal(labels, idx))
    classed_feature = tf.gather(features, locs)
    batch_mean, batch_var = tf.nn.moments(classed_feature, [0], name=var_name)
    loss = tf.square(tf.sub(classed_feature, batch_mean))
    return loss
```

```python
def center_loss_inter(moments, size, var_name):
    mean_list = []
    for i in range (0, n_class):
        mean_list.append(moments[i][0])
    packed = tf.pack(mean_list)
    packed = tf.reshape(packed, [n_class, size])
    packed_mean, packed_var = tf.nn.moments(packed, [0], name = var_name)
    loss = tf.square(tf.sub(packed, packed_mean))
    return loss
```

```python
def clac_layerwise_loss(moments, size, var_name):
    var_list = []
    for i in range (0, n_class):
        var_list.append(moments[i][1])
    packed_intra = tf.pack(var_list)  
    packed_intra = tf.reshape(packed_intra, [n_class, size])
    intra_var_loss = packed_intra
    #intra_var_loss = tf.reduce_sum(packed_intra)
    mean_list = []
    for i in range (0, n_class):
        mean_list.append(moments[i][0])
    packed = tf.pack(mean_list)
    
    packed = tf.reshape(packed, [n_class, size])
    packed_mean, packed_var = tf.nn.moments(packed, [0,1], name = var_name)
    #print packed_var.get_shape()
    #inter_var_loss = tf.reduce_sum(packed_var)
    inter_var_loss = packed_var
    #inter_var_loss = tf.square(tf.sub(inter_var_loss, 150))
    #layer_loss = intra_var_loss / inter_var_loss
    return intra_var_loss, inter_var_loss
    #return intra_var_loss + 1/inter_var_loss
```

```python
colors = cm.rainbow(np.linspace(0, 1, 10))
def make_ellipses(gmm, ax):
    for n, color in enumerate(colors):
        covariances = gmm.covariances_[n][:2, :2]

        v, w = np.linalg.eigh(covariances)
        u = w[0] / np.linalg.norm(w[0])
        angle = np.arctan2(u[1], u[0])
        angle = 180 * angle / np.pi  # convert to degrees
        v = 2. * np.sqrt(2.) * np.sqrt(v)
        ell = mpl.patches.Ellipse(gmm.means_[n, :2], v[0], v[1],
                                  180 + angle, color=color)
        ell.set_clip_box(ax.bbox)
        ell.set_alpha(0.5)
        ax.add_artist(ell)
        
n_classes = 10

# Try GMMs using different types of covariances.
estimators = dict((cov_type, GaussianMixture(n_components=n_classes,
                   covariance_type=cov_type, max_iter=20, random_state=0))
                  for cov_type in ['full'])

n_estimators = len(estimators)

plt.figure(figsize=(3 * n_estimators // 2, 6))
plt.subplots_adjust(bottom=.01, top=0.95, hspace=.15, wspace=.05,
                    left=.01, right=.99)
```

```python
def plot_results(X, Y_, means, covariances, index, title):
    splot = plt.subplot(2, 1, 1 + index)
    for i, (mean, covar, color) in enumerate(zip(
            means, covariances, colors)):
        v, w = linalg.eigh(covar)
        v = 2. * np.sqrt(2.) * np.sqrt(v)
        u = w[0] / linalg.norm(w[0])
        # as the DP will not use every component it has access to
        # unless it needs it, we shouldn't plot the redundant
        # components.
        if not np.any(Y_ == i):
            continue
        plt.scatter(X[Y_ == i, 0], X[Y_ == i, 1], .8, color=color)

        # Plot an ellipse to show the Gaussian component
        angle = np.arctan(u[1] / u[0])
        angle = 180. * angle / np.pi  # convert to degrees
        ell = mpl.patches.Ellipse(mean, v[0], v[1], 180. + angle, color=color)
        ell.set_clip_box(splot.bbox)
        ell.set_alpha(0.5)
        splot.add_artist(ell)

    #plt.xlim(-9., 5.)
    #plt.ylim(-3., 6.)
    #plt.xticks(())
    #plt.yticks(())
    plt.title(title)
```
